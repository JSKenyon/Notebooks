```python
import numpy as np
import pyrap.tables
import pylab as pb

%matplotlib inline
```

```python
ms = pyrap.tables.table("/home/jonathan/Notebooks/KAT7_1445_1x16_12h.ms")

uvw = ms.getcol("UVW")

A1 = ms.getcol("ANTENNA1")
A2 = ms.getcol("ANTENNA2")
```

```python
pb.plot(uvw[:,0],uvw[:,1], ".b",-uvw[:,0],-uvw[:,1], ".g", markersize=1)

pb.axis('image')

pb.xlabel("u(m)")
pb.ylabel("v(m)")

pb.show()
```

```python
for j in range(np.max(A1)+1):
	for k in range(j+1,np.max(A2)+1):														
		uvloc = (A1==j)&(A2==k) 
		u = uvw[uvloc,0]
		v = uvw[uvloc,1]
		pb.plot(u, v, color="b")
		pb.plot(-u, -v, color="g")
        
pb.axis('image')

pb.xlabel("u(m)")
pb.ylabel("v(m)")

pb.show()
```
