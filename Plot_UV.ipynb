```python
import numpy as np
import pyrap.tables
import pylab as pb

%matplotlib inline
```

```python
ms = pyrap.tables.table("/home/jonathan/Notebooks/KAT7_1445_1x16_12h.ms")

uv = ms.getcol("UVW")

A1 = ms.getcol("ANTENNA1")
A2 = ms.getcol("ANTENNA2")
```

```python
for j in range(np.max(A1)+1):
	for k in range(j+1,np.max(A2)+1):																											
		uvloc = (A1==j)&(A2==k) 
		u = uv[uvloc,0]
		v = uv[uvloc,1]
		pb.plot(u,v,color="b")
		pb.plot(-u,-v,color="g")
        
pb.axis('image')

pb.xlabel(r'u ($\lambda$)')
pb.ylabel(r'v ($\lambda$)')

pb.show()
```
